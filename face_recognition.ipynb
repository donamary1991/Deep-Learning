Why is face_location needed?

    Preprocessing for Recognition – Before recognizing a face, the model needs to detect and extract the face region.
    Improves Accuracy – Helps focus only on the face, ignoring background noise.
    Speeds Up Recognition – Instead of processing the whole image, it narrows down the area for recognition.

```python
def faceencoding(imgs):
    encodinglist=[]
    for img in imgs:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img)
        encode=fr.face_encodings(img,face_in_frame)[0]
        encodinglist.append(encode)
    return encodinglist

```

```python
encodelist_knownfaces=faceencoding(imgs)
print(encodelist_knownfaces)
```

```python
video=cv2.VideoCapture(0)
while True:
    sucess,frame=video.read()
    frames=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(frames)
    # print(face_in_frame)
    encoded_face=fr.face_encodings(frame,face_in_frame)
    # print(encoded_face)
    
    for enc_face,faceloc in zip(encoded_face,face_in_frame):
        matches=fr.compare_faces(encodelist_knownfaces,enc_face)
        print(matches)
        
        face_dist=fr.face_distance(encodelist_knownfaces,enc_face)
        print(face_dist)
        
        matchindex=np.argmin(face_dist)
        print(matchindex)
        
        if matches[matchindex]:
            name=class_name[matchindex]
            print(name)
        else:
            name="unknown"
            print(name)
        y1,x2,y2,x1=faceloc
        cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,0),3)
        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
            
        
            
    cv2.imshow("Face",frame)
    if cv2.waitKey(1)& 0XFF==ord('q'): 
        break
    
video.release()
cv2.destroyAllWindows()
# empty face=empty array
# single face=single_array
# two face=two array
# encoded_faces==[0] can be neglected since all faces need to be detected
# performance increase== face distance calculation
# compare training image with captured image==similar image==distance minimum

    
    
```
