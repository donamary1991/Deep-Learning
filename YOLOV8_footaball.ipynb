```python
import os
import shutil
# file handling--better performance--files copy,remove
import random
from tqdm.notebook import tqdm  #taqaddum==>progress
```

```python
!pwd
```

```python
# . means current directory
train_img_path="./yolo_data/images/train/"
train_label_path="./yolo_data/labels/train/"
val_img_path="./yolo_data/images/val/"
val_label_path="./yolo_data/labels/val/"
test_path='./yolo_data/test'
```

```python
random.seed(42)
```

```python
# splitting in to validation and training data
def train_test_split(path1,path2,split):
  # files=list(set([name for name in os.listdir(path)]))
  files=[name[:-4] for name in os.listdir(path1)]
  # remove'jpg'
  # print(files)
  random.shuffle(files)
  test_size=int(len(files)*split)
  train_size=len(files)-test_size

  # creation of directories
  os.makedirs(train_img_path,exist_ok=True)
  os.makedirs(train_label_path,exist_ok=True)
  os.makedirs(val_img_path,exist_ok=True)
  os.makedirs(val_label_path,exist_ok=True)


  for file in tqdm(files[:train_size]):
    # shutil.copy2(path+file+'.jpg',f"{train_img_path}/"+file+".jpg")
    shutil.copy2(path1+file+'.jpg',f"{train_img_path}/"+file+".jpg")
  #   # shutil.copy2(path+file+'.txt',f"{train_label_path}/"+file+".txt")
    shutil.copy2(path2+file+'.txt',f"{train_label_path}/"+file+".txt")

  for file in tqdm(files[train_size:]):
    shutil.copy2(path1+file+'.jpg',f"{val_img_path}/"+file+".jpg")
  #   shutil.copy2(path1+file,f"{val_img_path}/"+file)
    shutil.copy2(path2+file+'.txt',f"{val_label_path}/"+file+".txt")
  #   shutil.copy2


  # # copying files
  # for file in files
  #  print(file)


# each file has its jpg and text file hence took unique names
```
