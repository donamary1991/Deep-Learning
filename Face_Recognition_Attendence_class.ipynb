```python
import pandas as pd
import numpy as np
import cmake
import dlib
import face_recognition as fr
import os
import cv2
import csv
from datetime import datetime
```

```python
path='class_list'
class_path=os.listdir(path)
print(class_path)
```

```python
std_images=[]
std_name=[]
for img in class_path:
    img_path=os.path.join(path,img)
    # print(img_path)
    img_read=cv2.imread(img_path)
    # print(img_read)
    std_images.append(img_read)
    std_name.append(img[:-5])
print(std_name)
```

```python
# Face_Location & Encoding
def face_encoding(std_images):
    encoded_list=[]
    for img_pixels in std_images:
        img=cv2.cvtColor(img_pixels,cv2.COLOR_BGR2RGB)
        face_loc=fr.face_locations(img)
        face_enc=fr.face_encodings(img,face_loc)[0]
        encoded_list.append(face_enc)
    return encoded_list   
          
```

```python
encoded_student_faces=face_encoding(std_images)
print(encoded_student_faces)
```

```python
# Attendence Marking
video=cv2.VideoCapture(0)
# CSV file setup
csv_filename = "attendance_sheet.csv"
with open(csv_filename, mode='a', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Time"])
while True:
    success,frame=video.read()
    frames=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face_loc=fr.face_locations(frames)
    # print(face_loc)
    face_enc=fr.face_encodings(frame,face_loc)
    # print(face_enc)
    for faceenc,faceloc in zip(face_enc,face_loc):
        match_check=fr.compare_faces(encoded_student_faces,faceenc)
        print(match_check)
        dist_check=fr.face_distance(encoded_student_faces,faceenc)
        print(dist_check)
        match_index=np.argmin(dist_check)
        
        if match_check[match_index]:
            name=std_name[match_index]
            print(name)
            # Write to CSV file
            with open(csv_filename, mode='a', newline='') as file:
                writer = csv.writer(file)
                writer.writerow([name, datetime.now().strftime("%Y-%m-%d %H:%M:%S")])
                
        else:
            name="unkown"
            print(name)
        y1,x2,y2,x1=faceloc
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)
        
    cv2.imshow("Face",frame)
    if cv2.waitKey(1)& 0XFF==ord('q'): 
        break
video.release()
cv2.destroyAllWindows()
    
```
