```python
# OR
import ultralytics
from google.colab.patches import cv2_imshow
import cv2
import os

# Load the YOLOv8 model
model = ultralytics.YOLO('yolov8n.pt')

# Perform prediction on an image
results = model.predict(source="/content/jeep-cat.jpg", save=True)  # Replace with your image path
# Display the results
# Get the latest run directory from 'runs/detect'
runs_dir = os.path.join('runs', 'detect')
latest_run = sorted(os.listdir(runs_dir), key=lambda f: os.path.getmtime(os.path.join(runs_dir, f)), reverse=True)[0]
save_dir = os.path.join(runs_dir, latest_run)

# Display the results
for img_path in os.listdir(save_dir):
    if img_path.endswith(('.jpg', '.png')):  # Check for image files
        img = cv2.imread(os.path.join(save_dir, img_path))
        cv2_imshow(img)
```
